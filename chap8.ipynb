## Virus Spreading Modeling

### SI Model
$$S(t+1) = \alpha S(t)^2 + (1-\alpha N -\beta)S(t) + \beta N$$

### Critical parameters case

```python
N = 1000
beta = 0.01
alpha = 0.00002
Ss = [500]
Is = [500]
Steps = 100
for _ in range(Steps):
    s = Ss[-1]
    s = alpha*s*s + (1-alpha*N-beta)*s + beta*N
    Ss.append(s)
    Is.append(N-s)

with plt.xkcd():
    plt.scatter([i for i in range(Steps+1)],Ss, label = 'susceptible')
    plt.scatter([i for i in range(Steps+1)],Is, label = 'infected')
    plt.title(r"SI model for high contagious disease with critical $\alpha$ & $\beta$ values")
    plt.xlabel("Days")
    plt.ylabel("Sizes of Groups")
    plt.legend()
```

### SIR model

```python
N = 1000
beta = 0.01
alpha = 0.0001
gamma = 0.0001
Ss = [500]
Is = [500]
Rs = [0]
Steps = 365*20
for _ in range(Steps):
    s, i, r = Ss[-1], Is[-1], Rs[-1]
    s_next = s - alpha*s*i + beta*i
    i_next = i + alpha*s*i - beta*i - gamma*i
    r_next = r + gamma*i
    Ss.append(s_next)
    Is.append(i_next)
    Rs.append(r_next)

with plt.xkcd():
    plt.scatter([i for i in range(Steps+1)],Ss, label = 'susceptible')
    plt.scatter([i for i in range(Steps+1)],Is, label = 'infected')
    plt.scatter([i for i in range(Steps+1)],Rs, label = 'recovered')
    plt.title(r"SIR model for small $\gamma$")
    plt.xlabel("Days")
    plt.ylabel("Sizes of Groups")
    plt.legend()
```

### Vaccination case

```python
N = 1000
beta = 0.01
alpha = 0.0001
gamma = 0.0001
mu = 0.05
Ss = [500]
Is = [500]
Rs = [0]
Steps = 100
for _ in range(Steps):
    s, i, r = Ss[-1], Is[-1], Rs[-1]
    s_next = s - alpha*s*i + beta*i - mu *s
    i_next = i + alpha*s*i - beta*i - gamma*i
    r_next = r + gamma*i + mu * s
    Ss.append(s_next)
    Is.append(i_next)
    Rs.append(r_next)

with plt.xkcd():
    plt.scatter([i for i in range(Steps+1)],Ss, label = 'susceptible')
    plt.scatter([i for i in range(Steps+1)],Is, label = 'infected')
    plt.scatter([i for i in range(Steps+1)],Rs, label = 'recovered')
    plt.title(r"SIR model for small $\gamma$ and large $\mu$")
    plt.xlabel("Days")
    plt.ylabel("Sizes of Groups")
    plt.legend()
```
